```python
import pandas as pd 

df = pd.read_csv('The_Cancer_data_1500_V2.csv')
df
```

```python
label = df['Diagnosis']
features = df.drop(columns=['Diagnosis'])
print(label)
features
```

```python
from sklearn.model_selection import train_test_split

traindata, testdata, train_labels, test_labels = train_test_split(features, 
                                                                  label,
                                                                  stratify=label,
                                                                  test_size=0.2,
                                                                  random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

model = gnb.fit(traindata, train_labels)
```

```python
prediction = gnb.predict(testdata)
prediction
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(test_labels, prediction)
print(f'Model accurate in {acc*100} %')
```

Experimenting with other naive bayeses

```python
from sklearn.naive_bayes import BernoulliNB, CategoricalNB, ComplementNB, MultinomialNB

bnb = BernoulliNB()
catnb = CategoricalNB()
comnb = ComplementNB()
mnb = MultinomialNB()

gaus_model = gnb.fit(traindata, train_labels)
bernoulli_model = bnb.fit(traindata, train_labels)
categorical_model = catnb.fit(traindata, train_labels)
complement_model = comnb.fit(traindata, train_labels)
multinomial_model = mnb.fit(traindata, train_labels)

gauss_prediction = gaus_model.predict(testdata)
# print(gauss_prediction)
# model_list = [bernoulli_model, categorical_model, complement_model, multinomual_model]
bernoulli_prediction = bernoulli_model.predict(testdata)
categorical_prediction = categorical_model.predict(testdata)
complement_prediction = complement_model.predict(testdata)
multinomial_prediction = multinomial_model.predict(testdata) 

predictions_list = [gauss_prediction, bernoulli_prediction, categorical_prediction, complement_prediction, multinomial_prediction]
prediction_dict = {
    'Gauss': gauss_prediction,
    'Bernoulli': bernoulli_prediction,
    'Categorical': categorical_prediction,
    'Complement': complement_prediction,
    'Multinomial': multinomial_prediction
}

# test_acc = accuracy_score(test_labels, gauss_prediction)
# print(test_acc)
for name, prediction in prediction_dict.items():
    # print(name)
    acc = accuracy_score(test_labels, prediction)
    print(f'Accuracy for {name} is {acc*100}%')
```
